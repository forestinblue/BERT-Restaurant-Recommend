```python
from haversine import haversine
```

```python
#유사도 몇 점 이상만 거리 계산 이런식으로 -> 계산량 단축
#서울이면 경기도, 서울만  
#강원도이면 강원도, 서울, 경상도까지만 이런식으로 -> 계산량 단축

Pin_restaurant_lon =mealion_review_copy.iloc[index]['lon']
Pin_restaurant_lat =mealion_review_copy.iloc[index]['lat']
Pin_restaurant = (Pin_restaurant_lat, Pin_restaurant_lon)  #Latitude, Longitude


recommander_restaurant_dataframe= pd.DataFrame({'feed_id':[], 'feed_distance':[],'feed_index':[], 'feed_similarity':[], 'restaurant_name':[], 'restaurant_address':[]})

# 거리 계산
for i in range(len(mealion_review_copy)):
    review_similarity = restaurant_recommendation[i][1]
    if review_similarity >0.9835:
   
        recommander_restaurant_lon = mealion_review_copy.iloc[restaurant_recommendation[i][0]]['lon']
        recommander_restaurant_lat = mealion_review_copy.iloc[restaurant_recommendation[i][0]]['lat']
        recommander_restaurant = (recommander_restaurant_lat, recommander_restaurant_lon)
        review_restaurant_distance = haversine(Pin_restaurant, recommander_restaurant, unit = 'km')
        if review_restaurant_distance < 5:
            review_index = restaurant_recommendation[i][0]
            review_id = mealion_review_copy.iloc[review_index]['review_id']
            restaurant_name = mealion_review_copy.iloc[review_index]['restaurant_name']
            restaurant_address = mealion_review_copy.iloc[review_index]['address']
            review_similarity = restaurant_recommendation[i][1]
            recommander_restaurant_dataframe.loc[i] = [int(review_id),review_restaurant_distance,int(review_index), review_similarity,restaurant_name , restaurant_address]
```
